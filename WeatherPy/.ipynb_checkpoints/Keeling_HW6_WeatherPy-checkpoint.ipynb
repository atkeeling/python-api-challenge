Observations:
1. There are no cities less than -60 latitude
2. There are no cities greater than 80 latitude
3. At this time of year, the hottest temperatures are at 10 latitude and temps get cooler the further from 10 latitude
4. Most cities have an RH over 60% regardless of latitude
5. Cloudiness does not correlate with latitude
6. Wind does not correlate with latitude

```python
import json
import requests
import random
import pandas as pd
from citipy import citipy
from config import api_key
import matplotlib.pyplot as plt
```

```python
base_url = "http://api.openweathermap.org/data/2.5/weather?"
params = {
    "appid": api_key,
    "units":"imperial"
}
```

```python
weatherlist = []
for i in range(500):
    while True:
        try:    
            city = citipy.nearest_city(random.uniform(-90,90), random.uniform(-180,180))
            cityname = city.city_name
            countrycode = city.country_code
            params["q"]=cityname+", "+countrycode
            response = requests.get(base_url, params=params)
            response_json = response.json()
            data = {
                "name":cityname+", "+countrycode,
                "lat":response_json["coord"]["lat"],
                "lon":response_json["coord"]["lon"],
                "temp":response_json["main"]["temp"],
                "humidity":response_json["main"]["humidity"],
                "wind":response_json["wind"]["speed"],
                "cloudiness":response_json["clouds"]["all"]
                }
        except KeyError:
            continue
        break
    print(f"Count: {i+1}")
    print(f"City name: {cityname.title()}, {countrycode.upper()}")
    print(response.url)
    weatherlist.append(data)
```

```python
df = pd.DataFrame(weatherlist)
df.to_csv("city_weather.csv")
df.head()
```

```python
plt.scatter(df["lat"], df["temp"], marker="o", facecolors="red", edgecolors="black",alpha=0.75)
plt.title("Temperature vs. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.xlim(-90,90)
plt.grid()
plt.show()
plt.savefig("temp_vs_lat.png")
```

```python
plt.scatter(df["lat"], df["humidity"], marker="o", facecolors="red", edgecolors="black",alpha=0.75)
plt.title("Relative Humidity vs. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Relative Humidity")
plt.xlim(-90,90)
plt.grid()
plt.show()
plt.savefig("humid_vs_lat.png")
```

```python
plt.scatter(df["lat"], df["cloudiness"], marker="o", facecolors="red", edgecolors="black",alpha=0.75)
plt.title("Cloudiness (%) vs. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.xlim(-90,90)
plt.grid()
plt.show()
plt.savefig("clouds_vs_lat.png")
```

```python
plt.scatter(df["lat"], df["wind"], marker="o", facecolors="red", edgecolors="black",alpha=0.75)
plt.title("Wind Speed (mph) vs. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.xlim(-90,90)
plt.grid()
plt.show()
plt.savefig("wind_vs_lat.png")
```
